```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f_path = "./outputs/output_2.txt"
f = open(f_path)
lines = f.readlines()
```

```python
Y = []
for l in lines:
    arr = l.split()
    if len(arr) == 5:
        Y.append(float(arr[-1]))

plt.plot(Y)
plt.ylabel("loss")
plt.show()
```

```python
Acc = []
for l in lines:
    arr = l.split()
    if (arr[0] == 'Accuracy'):
        Acc.append(float(arr[1]))

plt.plot(Acc)
plt.ylabel("Accuracy")
plt.show()
```
